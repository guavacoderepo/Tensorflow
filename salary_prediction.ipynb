```python
# import all neccessary libs

import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
```

```python
# import salary dataset 

salary = pd.read_csv("Salary Data.csv")
salary = salary.drop("Job Title", axis=1)

salary
```

```python
# get out dependent and indepenent datasets
x = salary.drop("Salary", axis=1)
y = salary["Salary"]

x.head()
```

```python
# normalize our dataset
tf.random.set_seed(42)
ct = make_column_transformer(
    (MinMaxScaler(), ["Age", "Years of Experience"]),
    (OneHotEncoder(handle_unknown='ignore'), ["Gender", "Education Level"])
)

# split our dataset to test and train
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)

# fit the column transformer to our training dataset
ct.fit(x_train)

# tranforming trainig and test with normaliztion and onhotencoder
x_train_normalize = ct.transform(x_train)
x_test_normalize = ct.transform(x_test)

x_train_normalize[0]
```

```python
# let's check our shape
x_train.shape, x_train_normalize.shape

x_train_normalize
```

```python
# let's build neural network to feed our data 

# build our model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

# compile our model

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["mae"]
              )

# fit our model

model.fit(x_train_normalize, y_train, epochs=100)

```
